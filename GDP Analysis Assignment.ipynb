```python
#Reading the data from delimited file into the dataframe

#Function to find the file encoding so that we dont face encoding issue while reading the CSV files
def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc

file_encodingIA = find_encoding('C:\\Users\Ashutosh\\Downloads\\Data_IA.csv')
#print(file_encoding)

#Reading and storing the file in data-frame
df_GSDP = pd.read_csv("C:\\Users\Ashutosh\\Downloads\\Data_IA.csv",encoding=file_encodingIA)

#Displaying contents of dataframe for checking its values
df_GSDP
```

```python
# Assignment Part 1B:

# Ignoring the states WestBengal and J&K since the data-set is not available for these two states

# Filtering out the Union Territories (Delhi, Chandigarh, Andaman and Nicobar Islands,Puducherry etc.) for further 
# analysis since they are governed directly by the centre, not state governments

#Importing libraries which will be needed for merging multiple data-sets:
import os
import glob

# the path to the states csv file directory
mycsvdir = "C:\\Users\\Ashutosh\\Downloads\\Eligible_states"

# get all the csv files in that directory:
csvfiles = glob.glob(os.path.join(mycsvdir, '*.csv'))

# loop through the files and read them in with pandas
dataframes = []  # a list to hold all the individual pandas DataFrames

for csvfile in csvfiles:

# Adding new columns for the state in the merged dataframe:
    state = csvfile.split('-')[1]
    
# Using the encoding for the file found earlier using the "find_encoding" function:     
    df = (pd.read_csv(csvfile,encoding="ISO-8859-1").assign(State=state))
    dataframes.append(df)

# concatenate them all together
allstate_dataframe = pd.concat(dataframes, ignore_index=False)

sequence = ["S.No.","State","Item","2011-12","2012-13","2013-14","2014-15","2015-16","2016-17"]
allstate_dataframe = allstate_dataframe.reindex(columns=sequence)

# print out to a new csv file
allstate_dataframe.to_csv('all_state.csv',index=False)

# Important : This CSV needs to be removed manually if we are doing a merge or running this code again as merge code will 
# look for all CSV in this path.
```

```python
# Assignment Part 1-B: percentage contribution of primary, secondary and tertiary sectors as a percentage of total 
# GDP for all the states

allstate_dataframe_pt = allstate_dataframe_new.pivot_table('2014-15','State','Item')

# Adding new columns for primary,secondary and tertiary percentage calculation:
allstate_dataframe_pt['Primary_%'] = (allstate_dataframe_pt['Primary']/allstate_dataframe_pt['Gross State Domestic Product'])*100
allstate_dataframe_pt['Secondary_%'] = (allstate_dataframe_pt['Secondary']/allstate_dataframe_pt['Gross State Domestic Product'])*100
allstate_dataframe_pt['Tertiary_%'] = (allstate_dataframe_pt['Tertiary']/allstate_dataframe_pt['Gross State Domestic Product'])*100

# Removing further columns which are not needed:
allstate_dataframe_pt = allstate_dataframe_pt.drop(["Gross State Domestic Product","Primary","Secondary","Tertiary"], axis=1)
```

```python
allstate_dataframe_pt.head()
```

```python
# Plotting a Stacked bar-chart to represent the percentage contribution of primary, secondary and tertiary sectors as a
# percentage of total GDP for all the states.

Primary = allstate_dataframe_pt['Primary_%']

Secondary = allstate_dataframe_pt['Secondary_%']

Tertiary = allstate_dataframe_pt['Tertiary_%']

###Doing change:
allstate_dataframe_pt.reset_index(inplace=True)

States = allstate_dataframe_pt['State'] 

plt.figure(figsize=(14,8))

p1 = plt.bar(States, Primary)

p2 = plt.bar(States, Secondary, bottom=Primary)

p3 = plt.bar(States, Tertiary, bottom=np.array(Primary)+np.array(Secondary))

plt.ylabel('Total GSDP Percentage (%)')

plt.title('Percentage Contribution of Each sector to GSDP')

plt.xticks(States,rotation=90)

plt.yticks(np.arange(0, 110, 10));
plt.xlabel('States Of India')

plt.legend((p1[0], p2[0], p3[0]), ('Primary', 'Secondary', 'Tertiary'))

plt.show()
```
